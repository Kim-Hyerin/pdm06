### [도전1.] 균일분포와 정규분포를 비교하는 그림을 그리시오.
- subplot 이용
- style(제목, x-축 레이블, y-축 레이블, ...)

> [hint] 검색: "How to add title to subplots in Matplotlib?"

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
plt.hist(np.random.rand(1000),color='skyblue') #균일분포
```

```
plt.hist(np.random.randn(1000),color='yellowgreen') # 정규분포
```

### subplot(121), subplot(122)

```
fg = plt.figure(figsize=(10,5))

#균일분포
sp1 = fg.add_subplot(121)
sp1.hist(np.random.rand(1000),color='skyblue')
sp1.title.set_text('subplot121')
sp1.set(xlabel = 'x-label', ylabel = 'y-label')

#정규분포
sp2 = fg.add_subplot(122)
sp2.hist(np.random.randn(1000),color='yellowgreen')
sp2.title.set_text('subplot122')
sp2.set(xlabel = 'x-label', ylabel = 'y-label')

plt.show()
```

### [도전 2.] 0~9 사이의 무작위 정수 1000개를 만들어서 각 수의 빈도에 대한 히스토그램, 막대그래프, 꺽은선 그래프를 그리고 비교하시오.
- subplot 이용
- style(제목, x-축 레이블, y-축 레이블, ...)



```
n1000 = np.random.randint(0,10,1000)
```

```
value, count= np.unique(n1000, return_counts=True)
```

```
value
```

```
count
```

```
plt.hist(n1000)
plt.show()
```

```
fg = plt.figure(figsize=(15,5))

#히스토그램
sp1 = fg.add_subplot(131)
sp1.hist(count,color='skyblue')
sp1.title.set_text('Histogram')
sp1.set(xlabel = 'x-label', ylabel = 'y-label')

#막대그래프
sp2 = fg.add_subplot(132)
sp2.bar(value,count,color='yellowgreen')
sp2.title.set_text('Bar Graph')
sp2.set(xlabel = 'x-label', ylabel = 'y-label')

#꺾은선그래프
sp3 = fg.add_subplot(133)
sp3.plot(value,count,color='orange')
sp3.title.set_text('Line Graph')
sp3.set(xlabel = 'x-label', ylabel = 'y-label')

plt.show()
```
